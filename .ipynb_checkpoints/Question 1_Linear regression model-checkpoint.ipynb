```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
#Loading the csv file

data = pd.read_csv('advertising.csv')
print(data.head())
```

```python
# linear regression analysis using the TV column as the independent variable and Sales as the dependent variable

X = data['TV'].values.reshape(-1, 1)  # Independent variable (TV)
y = data['Sales'].values.reshape(-1, 1)  # Dependent variable (Sales)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Fitting the linear regression model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Calculate and report the regression equation (line of best fit) in the form of y = mx + b

m = model.coef_[0][0]  # Slope
b = model.intercept_[0]  # Intercept
print(f"Regression Equation: y = {m}x + {b}")
```

```python
# Plot a scatter plot of the TV and Sales data points and overlay the regression line
plt.scatter(X_test, y_test, color='red', label='Actual Data')
plt.plot(X_test, model.predict(X_test), color='blue', label='Regression Line')
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Advertising Spending ($)')
plt.ylabel('Sales ($)')
plt.legend()
plt.show()
```

```python
#Calculate and report the R-squared (R2) value

y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)
print(f"R-squared (R2) Value: {r_squared}")
```

```python
# Calculate and report the Mean Absolute Error (MAE) as an accuracy metric
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
```
